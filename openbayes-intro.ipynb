# Jupyter 编辑器

Jupyter Lab 是一个交互式的开发运行环境([Jupyter Lab](https://blog.jupyter.org/jupyterlab-is-ready-for-users-5a6f039b8906))。 你可以在这里运行 Python 脚本、其他的 Jupyter Notebook 以及其他的 Linux 程序。 `/output` 或者 `/openbayes/home` 目录中的所有文件都会被保存。



<br><br>

## 基本操作

### 停止当前的 Jupyter 编辑器

你可以在顶栏的状态栏中看到当前的状态。如果你想停止当前的编辑器，点击`停止`按钮。所有在 `/output` 目录中的文件都会被保存下来。 另外的一个路径是 `/openbayes/home`，两个路径实际上指向同一个目录。



*注意: 记得在停止工作区之前保存你的所有 Notebook。 停止后当前容器内的所有程序都会停止运行。*

### 收费 

❗️ **只要编辑器还在运行，就会被计费。** 如果你暂时不再需要使用，记得关闭编辑器。

### 恢复 Jupyter 编辑器

编辑器停止后，如果希望继续执行，在容器的`执行记录`右边的菜单中，选择`自此继续执行`。



<br><br>

## 代码/数据


绑定的代码和绑定的输出都放在了 `/output` 或者 `/openbayes/home` 中，两个路径实际指向同一个目录。

❗️**重要 - 数据请存放在如上所说的目录中**: 只有这个目录中的文件会在关闭编辑器后被保存下来。

<br/>

## 使用终端

你可以在编辑器中打开 Bash 终端来运行、调试代码，或者查看你的文件。点击左侧边栏上方的 *+* 按钮然后点击 *Terminal* 按钮来打开终端。



<br/>

## 编辑器空闲时的行为

当编辑器持续一段时间无交互操作且占用非常少的 CPU 时，系统会认定编辑器处于闲置状态。在闲置状态持续一段时间后，系统会发送提醒提示编辑器闲置。如果编辑器仍然持续闲置，会被关闭。
这个机制是为了节省费用。如果你不希望编辑器在闲置时被关闭，可以在容器设置中调整超时时长，或者禁用闲置关闭功能。
